## Mongo Queries

```python
#bills.find is how you query, insert_many is how you add data, count_documents counts the number of documents
bills.count_documents({})
```

```python
bills.find_one({})
#pulls just one record - the first one - the most recent one 
```

```python
#bills.find({filter rows}, {filter columns})
#two dictionaries in bills - first filters rows, second filters the columns
```

```python
myquery = bills.find({'introduced_date': '2022-11-16'}, 
                     {'_id':0 , 'short_title':1, 'sponsor_name':1, 'introduced_date':1})
#id is the only column you have to manually turn off, 1 means include

pd.DataFrame.from_records(loads(dumps(myquery)))
#just pull bills sponsored on 2022-11-16
#dumps turns mongo output to string and loads turns it into dictionary, pd.DataFrame.from_records turns it into DB
```

```python
#Find all enacted laws
myquery = bills.find({'enacted': {'$ne' : None} }, 
                     {'_id':0, 'enacted':1, 'short_title':1, 'sponsor_name':1})
#id is the only column you have to manually turn off, 1 means include

laws = pd.DataFrame.from_records(loads(dumps(myquery)))
#just pull bills that were enacted
#dumps turns mongo output to string and loads turns it into dictionary, pd.DataFrame.from_records turns it into DB
```

```python
#who got the most laws passed
laws['sponsor_name'].value_counts()
```

```python
#Build search engine and find all bills that had at least 5 cosponsors and sort those by most relevant to ukraine
```

```python
bills.create_index([('short_title', 'text')])
#intitilizing search engine - give name of feature you want to do search on and text is what you are searching for 
```

```python
myquery = bills.find({'cosponsors': {'$gte': 5}, '$text': {'$search': 'ukraine', '$caseSensitive': False}},
                     {'_id':0, 'score': {'$meta': 'textScore'}, 'cosponsors': 1, 'short_title':1})
myquery.sort([('score', {'$meta': 'textScore'})])
pd.DataFrame.from_records(loads(dumps(myquery)))
```
